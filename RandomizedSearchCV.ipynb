```python
# importing sklearn models and datasets
from sklearn.model_selection import RandomizedSearchCV
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier as knn
```

```python
# dictonary that holds all the details of the process
Models = {
    
    # Model
    "svm" : {
        
        # Applying Model
        "Name" : svm.SVC(),
        
        # list of hyper parameters
        "HyperParams" : {
            "C" : [1.0, 3.0, 5.0, 7.0, 9.5, 10.0],
            "kernel" : ["linear", "rbf"],
            "gamma" : ["auto", "scale"]
        }
    },
    
    "RFC" : {
        "Name" : RandomForestClassifier(),
        "HyperParams" : {
            "n_estimators" : [5, 10, 15, 20, 25, 30, 45],
        }
    },
    
    "LR" : {
        "Name" : LogisticRegression(multi_class="auto"),
        'HyperParams' : {
            "C" : [1, 5, 10, 15, 20],
            "max_iter" : [100, 500, 1000, 800, 350, 1200],
            "solver" : ["newton-cg", "lbfgs", "liblinear", "sag", "saga"]
        }
    },
    
    "MultinominalNB" : {
        "Name" : MultinomialNB(alpha=1.0, fit_prior=True, class_prior=None),
        "HyperParams" : {}
    },
    
    "DecisionTree": {
        "Name" : DecisionTreeClassifier(),
        "HyperParams" : {}
    },
    
    "GaussinaNB" : {
        "Name" : GaussianNB(priors=None, var_smoothing=1e-09),
        "HyperParams" : {}
    },
    
    "KNN" : {
        "Name" : knn(),
        "HyperParams" : {
            "n_neighbors" : [4, 5, 6, 7, 8]
        },
    }
}
```

```python
irisdata = datasets.load_iris()
scores = []

for ModelName, ModelDetails in Models.items():
    Classifier = RandomizedSearchCV(ModelDetails["Name"], ModelDetails["HyperParams"], 
                                    n_iter = 30, cv = 4, return_train_score = False )
    Classifier.fit(irisdata.data, irisdata.target)
    
    scores.append({
        "Model" : ModelName,
        "BestScore" : Classifier.best_score_,
        "BestParams" : Classifier.best_params_
    })
```

```python
scores = sorted(scores, key = lambda x: x["BestScore"], reverse = True)

import pandas as pd
DataFrame = pd.DataFrame(scores, columns = ["Model", "BestScore", "BestParams"])
```
